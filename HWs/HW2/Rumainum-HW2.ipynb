```python
#Problem 3
#Define a function expo that performs this task, and state its computational 
#complexity using big-O notation. The first argument of this function is 
#the number,and the second argument is the exponent (non-negative numbers only).

def expo(num, expont):
    # initialize result
    result = num
    # calculate exponent
    if expont > 0:
        for i in range(1,expont):
            result *= num
    # any number to the power of 0 is 1
    else:
        return 1
    return result

# The big O-notation for this will be big-O(n) where n is the value of
# the exponent 

#initialize input as false input
isValid_input = False
while (isValid_input == False):
    #get input number and exponent
    inputNum = int(input("Enter number: "))
    inputExpo = int(input("Enter exponent: "))
    
    #check if it's greater than 0
    if (inputNum >= 0 and inputExpo >= 0):
        #update input to True to break the loop
        isValid_input = True
    else: #error message
        print("Error! Enter input >= 0!")

#print the result
print("The result is: ", expo(inputNum, inputExpo))
```

```python
# Problem 4
#Write the Fibonacci sequence as a linear (non-recursive) algorithm. Let the 
#function take in the desired term in the Fibonacci sequence, and 
#return the appropriate number.

def fibonacci(n):
    # initialize the fibonacci sequence
    result = 1
    # first two results on the sequence 
    a = 1
    b = 1
    
    # calculate the terms for n > 2
    if (n > 2):
        for i in range (2, n):
            #add both previous terms
            result = a + b
            #update a and b numbers accordingly
            #a->b and b->result
            a = b
            b = result
            
    #return result
    return result

#initialize input as false input
isValid_input = False
while (isValid_input == False):
    #get user's input number for which term of Fibonnaci sequence
    inputNum = int(input("Enter which term of Fibonnaci sequence: "))
    
    #check if it's greater than 0
    if (inputNum > 0):
        #update input to True to break the loop
        isValid_input = True
    else: #error message
        print("Error! Enter input > 0!")

# print result of the term
print("The Fibonnaci number for that term: ", fibonacci(inputNum))
```
