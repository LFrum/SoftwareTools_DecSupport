```python
#Problem 1
#Reverse a list

def reverseList(inputList):
    #create count from the length of input list
    count = len(inputList) - 1
    
    #create empty reverse list
    revList = []
    
    while count >= 0:
        #append to new list
        revList.append(inputList[count])
        #decrement count
        count = count - 1
    
    return revList

originalList = [1,2,3,4,5]
backwardList = reverseList(originalList)

print("Initial List: ", originalList)
print("Reverse List: ", backwardList)

#This method has computation complexity of big-O(n)
#since it'll always go through n-number of items in the list
```

```python
#Problem 2
#Medical Equipment
#Nodes = ["F1","F2","WH1", "WH2","WS1", "WS2", "WS3","WS4"]
F =["F1","F2"]
WH =["WH1", "WH2"]
WS =["WS1", "WS2","WS3", "WS4"]
supply={}
supply["F1"]=400
supply["F2"]=250

demand={}
demand["WS1"] = 200
demand["WS2"] = 100
demand["WS3"] = 150
demand["WS4"] = 200

#set cost to ridiciously high number 
#for non-existing arc
Cost_FWH = {
  ("F1","WH1") : 40,
  ("F1","WH2") : 35,
   ("F2","WH1"): 100000,
   ("F2","WH2"): 25
}

Cost_WHWS = {
    ("WH1","WS1") : 60,
    ("WH1","WS2") : 35,
    ("WH1","WS3") : 10000,
    ("WH1","WS4") : 10000,
    ("WH2","WS1") : 10000,
    ("WH2","WS2") : 55,
    ("WH2","WS3") : 50,
    ("WH2","WS4") : 65
}
```

```python
from gurobipy import * 
model = Model("MedEquipment")
    
X={}
Y={}

for i in F: 
    for j in WH:
        X[i,j] = model.addVar(vtype=GRB.INTEGER,lb=0,ub=GRB.INFINITY)

for j in WH: 
    for k in WS:
        Y[j,k] = model.addVar(vtype=GRB.INTEGER,lb=0,ub=GRB.INFINITY)

model.modelSense = GRB.MINIMIZE                   
model.update()
```

```python
#demand constraints
for k in WS:    
    model.addConstr(quicksum(Y[j,k] for j in WH) == demand[k])   

# maximum capacity 
model.addConstr(X["F1","WH1"] <= 250) 

for i in F:    
    model.addConstr(quicksum(X[i,j] for j in WH) <= supply[i])  

# Equilibrium H
for j in WH:
    model.addConstr(quicksum(X[i,j] for i in F) - quicksum(Y[j,k] for k in WS) == 0) 


objective = quicksum(Cost_FWH[i,j]*X[i,j] for j in WH for i in F) + quicksum(Cost_WHWS[j,k]*Y[j,k] for k in WS for j in WH)

model.setObjective(objective)
```

```python
model.optimize()
```

```python
#Printing outputs
if model.status==GRB.OPTIMAL:
    print ("Optimal value:", model.objVal)
    print ("--- Quantity (Factory to Warehouse)---")
    for i in F: 
        for j in WH:
            print (i, j, X[i,j].x)
    print ("--- Quantity (Warehouse to Wholesaler)---")
    for j in WH: 
        for k in WS:
            print ( j, k, Y[j,k].x)   
```
