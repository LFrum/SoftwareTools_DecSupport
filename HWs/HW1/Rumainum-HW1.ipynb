```python
#Lince Rumainum
#ISE5123
#HW-1

#Problem 1
#Finding approximation of pi
#based on German mathematician Gottfried Leibniz's method

#initialize the summation of RHS value
sumRHS = 0

#initialize input as false input
isValid_input = False
while (isValid_input == False):
    #get input for how many iterations
    inputIter = int(input("Enter number of iteration to approximate pi's value: "))
    
    #check if it's greater than 0
    if (inputIter > 0):
        #update input to True to break the loop
        isValid_input = True
    else: #error message
        print("Error! Enter input > 0!")

# Calculate the RHS of Gottfried Leibniz's method
for i in list(range(1,inputIter+1)):
    #calculate the summation for the iteration(s)
    if (i > 1):
        sumRHS += (-1)**(i-1) * 1/(2*i-1)
    else:
        sumRHS += 1/i
    
# multiply the summation by 4 to get approximation of pi
approxPi = 4 * sumRHS

# print the pi value
print ("The approximate value of pi: ", approxPi)
```

```python
#Lince Rumainum
#ISE5123
#HW-1

#Problem 4
#The payroll department keeps a list of employee information for each pay period
#in a text file. The format of each line of the le is the following:
#< lastname > < department > < hourlywage > < hoursworked >

#Write a program that inputs a lename from the user and prints to the terminal a
#report of the wages paid to the employees for the given period. The report should
#be in tabular format with the appropriate header. Each line should contain an
#employee's name, their department, the hours worked, and the total wages paid
#for that period.

# open input file to read
inputFile = open("data.txt",'r')
# print header
print("   Last Name |  Department | Hours Worked | Total Wage Paid")


# go to each line of the input text file
for line in inputFile:
    # initialize data list
    data = []
    # strip each line on the file
    line = line.strip()    
    
    # check if whitespace or enter
    if (line != ""):      
        # split data word by word
        for words in line.split():
            # put each word into data list
            data.append(words)

        #put data accordingly
        lastName = str(data[0]) #last name
        dept = str(data[1]) #department
        hourlyWage = str(data[2]) #hourly wage
        hoursWorked = str(data[3]) #hours worked
    
        #calculate total wages
        totalWage = float(hourlyWage) * float(hoursWorked)
        
        # see length of each data for spacing on table
        lenName = len(lastName)
        lenDept = len(dept)
        lenHoursWorked = len(hoursWorked)        
        lenTotalWage = len(str(totalWage))
        
        # print data
        print(" "*(11-lenName), lastName,
              "| " + " "*(10-lenDept), dept,
              "| " + " "*(11-lenHoursWorked), hoursWorked,
              "| " + " "*(14-lenTotalWage), round(totalWage,2))
        
#close input file
inputFile.close()
    
```
