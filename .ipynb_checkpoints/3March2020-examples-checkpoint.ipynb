```python
#3 March 2020
#ISE-5123
#Lecture 13

#NETWORK: EXAMPLE 1
import networkx as nx
g = nx.Graph() #create an undirected graph
g.add_edge('a','b',weight=0.1)
g.add_edge('b','c',weight=1.5)
g.add_edge('a','c',weight=1.0)
g.add_edge('c','d',weight=2.2)
```

```python
import matplotlib.pyplot as plt

nx.draw(g)
plt.show()
```

```python
g.nodes()
```

```python
g.edges()
```

```python
g.degree('a')
```

```python
N,K = g.order(), g.size()
degrees = dict(g.degree())
avg_deg=sum(degrees.values())/float(len(degrees.values())) #for an undirected graph
aveg_deg= float(K/N)   # for a directed graph
print ("Nodes: ", N)
print ("Edges: ", K)
print ("Average degree: ", avg_deg)
```

```python
nx.shortest_path(g,'b','d')
```

```python
nx.shortest_path(g, source=None, target=None, weight=True)
```

```python
#NETWORK: EXAMPLE II
#Load the Hartford drug users network: 
#it is a directed graph with integers as nodes

#nx.DiGraph() --> DiGraph means Directed graph
#for undirected graph --> nx.Graph(), the same as example 1
hartford = nx.read_edgelist('hartford.txt', create_using=nx.DiGraph(),nodetype=int)
N,K = hartford.order(), hartford.size()  
print ("Nodes: ", N)
print ("Edges: ", K)
nx.draw(hartford)
plt.show() 
```

```python
aveg_deg = float(K/N)

print ("Average degree: ", aveg_deg)
```
